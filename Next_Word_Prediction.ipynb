```python
import pandas as pd
import os

# Dataset Path
path = r"C:\Users\digvijay\.cache\kagglehub\datasets\datatattle\covid-19-nlp-text-classification\versions\1"

# Load Dataset
df = pd.read_csv(os.path.join(path, "Corona_NLP_train.csv"), encoding="latin1")

# First Tweet
text = df["OriginalTweet"][0]

# Tokenization
tokens = text.split()

print(tokens)
```

```python
print("Total Tokens:", len(tokens))

seq_length = 5

X = []
Y = []

for i in range(len(tokens) - seq_length):
    X.append(tokens[i:i+seq_length])
    Y.append(tokens[i+seq_length])

print("Total Pairs:", len(X))

if len(X) > 0:
    print("First Input:", X[0])
    print("First Output:", Y[0])
else:
    print("Sequence length is too large.")
```

```python

# Create Vocabulary

vocab = sorted(set(tokens))

print("Vocabulary Size:", len(vocab))

print(vocab[:20])    # First 20 words
```

```python
X_encoded = []

for seq in X:
    X_encoded.append([word_to_index[word] for word in seq])

Y_encoded = []

for word in Y:
    Y_encoded.append(word_to_index[word])

print(X_encoded[:3])
print(Y_encoded[:3])
```

```python
# Create Vocabulary

vocab = sorted(set(tokens))

print("Vocabulary Size:", len(vocab))

print("\nFirst 20 Words:")
print(vocab[:20])
```

```python
word_to_index = {}

for index, word in enumerate(vocab):
    word_to_index[word] = index

print(word_to_index)
```

```python
index_to_word = {}

for index, word in enumerate(vocab):
    index_to_word[index] = word

print(index_to_word)
```

```python
X_encoded = []

for sequence in X:
    encoded_sequence = []

    for word in sequence:
        encoded_sequence.append(word_to_index[word])

    X_encoded.append(encoded_sequence)

print("First Encoded Input:")
print(X_encoded[0])
```

```python
Y_encoded = []

for word in Y:
    Y_encoded.append(word_to_index[word])

print("First Encoded Output:")
print(Y_encoded[0])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Create Model
model = Sequential()

# Embedding Layer
model.add(Embedding(
    input_dim=len(vocab),   # Vocabulary Size
    output_dim=100,         # Embedding Dimension
    input_length=seq_length # Input Sequence Length
))

# LSTM Layer
model.add(LSTM(150))

# Output Layer
model.add(Dense(len(vocab), activation='softmax'))

# Model Summary
model.summary()
```

```python
model.compile(
    loss='sparse_categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)
```
